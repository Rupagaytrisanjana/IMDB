#                    DATA ANALYSIS FOR IMDB MOVIE DATASET 

```python
#import libraries(these are the libraries i'm familiar with while performing EDA steps)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading data into dataframe(loaded it from kaggle)
data= pd.read_csv('IMDB-Movie-Data.csv')
```

```python
#display top 5 rows
data.head(10)
```

```python
#display last 5 rows (head and tail are used to know the stating entry and ending entry of a dataset here the starting entry is 0 and ending entry is 999 in total 1000 rows)
data.tail(10)
```

```python
#checking types of data(here we can see the types of data using dtypes)
data.dtypes
```

```python
#info of data(used to identify the data)
data.info()
```

```python
#knowing shape helps us to know no.of rows and columns
data.shape
```

```python
# Checking missing values
data.isnull()
```

```python
data.isnull().sum()
```

```python
#vizualize missing values using seaborn heatmap
sns.heatmap(data.isnull())
```

```python
#viewing missing values in percentages
per_missing=data.isnull().sum()*100 /len(data) # 13% and 7% is missing
per_missing                                   
```

```python
#drop missing values
data.dropna(axis=0)
```

```python
#dropping any duplicates
data=data.drop_duplicates()
data
```

```python
#displaying statistics
data.describe()
```

```python
#display title of movie having runtime>= 180 minutes
data[data['Runtime (Minutes)']>= 180]['Title']
```

```python
# we can always vizualize for better understanding
sns.barplot(x='Year',y='Votes',data=data)
```

```python
#we can make it neat 
sns.barplot(x='Year',y='Votes',data=data)
plt.title("votes by Year")
plt.show
```

```python
data.columns
```

```python
#which year,there was highest revenue
sns.barplot(x='Year',y='Revenue (Millions)',data=data)
plt.title("votes by Year")
plt.show
```

```python
#find avg rate of each director
data.groupby('Director')['Rating'].mean()
```
